```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv("./tips-data.csv")
sex_series = tips["sex"]
time_series = tips["time"]
smoker_series = tips["smoker"]
tip_series = tips["tip"]
total_bill_series = tips["total_bill"]

tips
```

```python
plt.figure(figsize=(5, 5))
plt.pie(sex_series.value_counts().head(3), labels=sex_series.value_counts().head(3).index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution Pie Chart')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(total_bill_series, kde=True, color="red")
plt.title('Histogram of Total Bill')
plt.xlabel("Total Bill")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(5, 5))
plt.pie(time_series.value_counts().head(3), labels=time_series.value_counts().head(3).index, autopct='%1.1f%%', startangle=140)
plt.title('Time Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 5))

sns.histplot(tip_series, kde=True, color='green')
plt.title('Histogram of Tip')
plt.xlabel('Tip')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='total_bill', data=tips)
plt.title('Bar Plot of Total Bill by Gender')
plt.ylabel("Total Bill")
plt.xlabel("Sex")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(total_bill_series, label='Total Bill')
plt.plot(tip_series, label='Tip')
plt.title('Line Chart of Total Bill and Tip')
plt.xlabel("Customer Count")
plt.ylabel("Amount (in $)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Gender')
plt.xlabel("Sex")
plt.ylabel("Total Bill")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.title('Linear Regression Plot of Total Bill vs Tip')
plt.show()
```

```python
df_corr = tips[['total_bill','tip', 'size']].corr()
sns.heatmap(df_corr,annot=True, cmap='PuOr')
plt.title("Total Bill, Tip and Size Heatmap")
plt.show()
```
